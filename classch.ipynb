```python
from tld import get_tld
import pandas as pd
import numpy as np
from collections import Counter
import math

def entropy(s):
    p, lns = Counter(s),float(len(s))
    return -sum(count/lns * math.log(count/lns,2)for count in p.values())
domaindata=pd.read_csv('./desktop/class1_19.csv')
entropydata=[]
length=[]
for i in mainurl:
    entropydata.append(entropy(i))
    length.append(len(i))
domaindata['entropy']=entropydata
domaindata['MainLength']=length
domaindata.to_csv('./desktop/class0.csv')
```

```python
import math
def entropy(string):
    prob = [ float(string.count(c)) / len(string) for c in dict.fromkeys(list(string)) ]
    entropy = - sum([ p * math.log(p) / math.log(2.0) for p in prob ])
    return entropy
entropy("heysl")
```
